```python
#libraries
import datetime as dt
import xarray as xr
import numpy as np
import fsspec
import s3fs
import os.path
import matplotlib.pyplot as plt

# make datasets display nicely
xr.set_options(display_style="html")  

#magic fncts #put static images of your plot embedded in the notebook
%matplotlib inline  
plt.rcParams['figure.figsize'] = 12, 6
%config InlineBackend.figure_format = 'retina' 
```

```python
%%time

file_location = 's3://ohw-bucket/goes_east'

ikey = fsspec.get_mapper(file_location,anon=False)

ds_e = xr.open_zarr(ikey) #,consolidated=True)

ds_e
```

```python
%%time

file_location = 's3://ohw-bucket/goes_west'

ikey = fsspec.get_mapper(file_location,anon=False)

ds_w = xr.open_zarr(ikey) #,consolidated=True)

ds_w
```

### Full Disc Image

```python
masked_e = ds_e.where(ds_e.DQF==0)
masked_e = masked_e.isel(time=slice(0,36))
masked_e = masked_e.SST.mean('time',skipna=True)



gs_subset = ds_e.sel(x=slice(-0.012,0.012),y=slice(0.12,0.085))  #reduce to GS region
masked_subset = gs_subset.SST.where(gs_subset.DQF==0)
masked_subset = masked_subset.isel(time=slice(0,36))
masked_subset = masked_subset.mean('time') 
```

```python
masked_w = ds_w.where(ds_w.DQF==0)
masked_w = masked_w.isel(time=slice(0,36))
masked_w = masked_w.SST.mean('time',skipna=True)
```

```python
plt.ioff()


fig, ax = plt.subplots(1,3, figsize=(20,6))



masked_e.plot(ax=ax[0], cmap='inferno', add_colorbar=False)
ax[0].axis('off')


masked_w.plot(ax=ax[1], cmap='inferno', add_colorbar=False)
ax[1].axis('off')

masked_subset.plot(ax=ax[2], cmap='inferno', add_colorbar=False)
ax[2].axis('off')

ax[0].set_title('GOES-East')
ax[1].set_title('GOES-West')
ax[2].set_title('Gulf Stream')

plt.savefig('data_overview.png')

plt.close(fig)
```
